```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
```

```python
directory="/Users/mdsanowarhossain/Documents/python/Deep_Learning/code_planet/potato_leaf"
category=["Early_blight","Healthy","Late_blight"]
```

```python
data=[]
for cat in category:
    newfolder=os.path.join(directory,cat)
    #print(newfolder)
    label=category.index(cat)
    
    for img in os.listdir(newfolder):
        img=os.path.join(newfolder,img)
        img_arr=cv2.imread(img)
        img_arr=cv2.resize(img_arr,(100,100))
        #print(img_arr)
        plt.imshow(img_arr)
        break
        data.append([img_arr,label])
```

```python
data
```

```python
#shuffle data as random wise
random.shuffle(data)
```

```python
#data split as dependent and independent
x=[]
y=[]
for feature, label in data:
    x.append(feature)
    y.append(label)
```
